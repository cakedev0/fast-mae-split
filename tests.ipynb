```python
from time import perf_counter as time

import numpy as np
import matplotlib.pyplot as plt

from lib.ds.segmenttree import WeightedSegmentTree
from lib.algo import compute_prefix_loss_heap, compute_prefix_loss_segmenttree
```

```python
n = 100

y = np.random.rand(n)
w = np.ones(n)
l = compute_prefix_loss_heap(y, w)
l_ = compute_prefix_loss_segmenttree(y, w)
assert np.allclose(l, l_)
```

```python
def rankdata(x):
    rank = np.empty(x.size, dtype=np.int32)
    rank[np.argsort(x)] = np.arange(0, rank.size)
    return rank
```

```python
def plot_execution_time_comparison(sample_data, max_log_n=6):
    ns = np.logspace(max_log_n - 2, max_log_n, num=20).astype(int)

    dt_heap = []
    dt_st = []

    for n in ns:
        y, w = sample_data(n)
        t = time()
        l = compute_prefix_loss_heap(y, w)
        dt_heap.append(time() - t)
        t = time()
        l_ = compute_prefix_loss_segmenttree(y, w)
        dt_st.append(time() - t)

    plt.plot(ns, dt_heap)
    plt.plot(ns, dt_st)
    plt.legend(['heap', 'segment-tree'])
    plt.xlabel('n')
    plt.ylabel('execution time')
    plt.title(sample_data.__name__)
    plt.show()
```

```python
def unit_weights(n):
    return np.random.rand(n), np.ones(n)

plot_execution_time_comparison(unit_weights)
```

```python
def uniform_weights(n):
    return np.random.rand(n), np.random.rand(n)

plot_execution_time_comparison(uniform_weights)
```

```python
def log_normal_weights(n):
    return np.random.rand(n), np.random.lognormal(size=n)

def exponential_weights(n):
    return np.random.rand(n), np.random.exponential(size=n)

plot_execution_time_comparison(log_normal_weights)
plot_execution_time_comparison(exponential_weights)
```

```python
def small_weights_around_median(n):
    y = np.random.rand(n)
    w = np.ones(n)
    w[(0.25 < y) & (y < 0.75)] = 0.01
    return y, w

plot_execution_time_comparison(small_weights_around_median)
```

```python
def worst_case_scenario(n):
    r = np.arange(n)
    y = np.power(-1.0005, r)
    w = np.power(1.0005, r)
    return y, w

plot_execution_time_comparison(worst_case_scenario, max_log_n=4.5)
```
